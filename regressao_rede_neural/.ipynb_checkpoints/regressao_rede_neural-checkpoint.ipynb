```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dados = pd.read_csv('./data/Bicicletas.csv')
dados.head()
```

```python
def plotar_eixos(x, y, labels):
    plt.scatter(x,y)
    plt.xlabel(labels[0])
    plt.ylabel(labels[1])
    
plotar_eixos(dados['temperatura'], 
             dados['bicicletas_alugadas'], 
             ['temperatura', 'quantidade_bicicletas_alugadas'])
```

```python
plotar_eixos(dados['clima'], 
             dados['bicicletas_alugadas'], 
             ['clima_categoria', 'quantidade_bicicletas_alugadas'])
```
